# P1: Boston Housing Price Prediction using Deep Neural Network

```python
# Step 1: Import Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# Step 2: Load Dataset
# boston_housing was removed from keras.datasets in newer versions.
# We load it directly from sklearn or a bundled numpy file instead.
from tensorflow.keras.datasets import boston_housing
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
# Step 3: Explore the Data
print("Training set shape:", x_train.shape)
print("Test set shape:", x_test.shape)
print("First 10 input samples:")
print(x_train[:10])
print("First 10 target values:")
print(y_train[:10])
```

```python
# Step 4: Feature Scaling (Normalization)
# Compute mean and std ONLY from training data to avoid data leakage
mean = x_train.mean(axis=0)
std = x_train.std(axis=0)

x_train = (x_train - mean) / std
x_test = (x_test - mean) / std  # Use training stats on test data
```

```python
# Step 5: Build Deep Neural Network Model
# FIX: Original code had a typo 'layers.Denslivation' — corrected to layers.Dense with proper activation
model = keras.Sequential([
    keras.Input(shape=(x_train.shape[1],)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)  # Output layer: single neuron for regression
])

model.summary()
```

```python
# Step 6: Compile Model
model.compile(
    optimizer='adam',
    loss='mse',        # Mean Squared Error for regression
    metrics=['mae']    # Mean Absolute Error for interpretability
)
```

```python
# Step 7: Train the Model
history = model.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.2  # 20% of training data used for validation
)
```

```python
# Step 8: Model Evaluation
test_loss, test_mae = model.evaluate(x_test, y_test)
print("Test Loss (MSE):", test_loss)
print("Test MAE:", test_mae)
```

```python
# Step 9: Prediction + Result Comparison
predictions = model.predict(x_test)
predictions = predictions.flatten()  # Convert 2D to 1D array

print("Predicted values:", predictions[:5])
print("Actual values:   ", y_test[:5])
```

```python
# Step 10: Plot Training History
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss (MSE) over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Train MAE')
plt.plot(history.history['val_mae'], label='Val MAE')
plt.title('MAE over Epochs')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Step 11: Visualise Predicted vs Actual
plt.figure(figsize=(8, 5))
plt.scatter(y_test, predictions, alpha=0.7, color='steelblue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Perfect prediction')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted House Prices')
plt.legend()
plt.show()
```
