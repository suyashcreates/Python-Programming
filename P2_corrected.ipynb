```python
# Step 2: Import Libraries and Load Dataset
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

print("Training data shape:", x_train.shape)  # (60000, 28, 28)
print("Testing data shape: ", x_test.shape)   # (10000, 28, 28)
```

```python
# Step 3: Show Sample Images
class_names = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
               'Sandal','Shirt','Sneaker','Bag','Ankle boot']

plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(class_names[y_train[i]])
    plt.axis('off')
plt.suptitle('Sample Images from Fashion MNIST')
plt.tight_layout()
plt.show()
```

```python
# Step 4: Normalize Data
# Pixel values are 0-255; divide by 255 to scale them to [0, 1]
x_train = x_train / 255.0
x_test  = x_test  / 255.0
```

```python
# Step 5: Build CNN Model
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Input(shape=(28, 28)),       # Input: 28x28 grayscale images

    layers.Reshape((28, 28, 1)),        # Add channel dimension -> (28, 28, 1)

    # First Conv block
    layers.Conv2D(32, (3, 3), activation='relu'),   # 32 filters, 3x3 kernel
    layers.MaxPooling2D((2, 2)),                     # Downsample by 2x

    # Second Conv block
    layers.Conv2D(64, (3, 3), activation='relu'),   # 64 filters, 3x3 kernel
    layers.MaxPooling2D((2, 2)),                     # Downsample by 2x

    # Fully Connected layers
    layers.Flatten(),                               # Flatten feature maps
    layers.Dense(128, activation='relu'),           # Hidden layer
    layers.Dense(10, activation='softmax')          # Output: 10 classes
])

model.summary()
```

```python
# Step 6: Compile Model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # Labels are integers (not one-hot)
    metrics=['accuracy']
)
```

```python
# Step 7: Train Model
# FIX: Original code trained twice (once without history, once with).
# Corrected to train ONCE and store history for plotting.
history = model.fit(
    x_train, y_train,
    epochs=5,
    validation_data=(x_test, y_test)
)
```

```python
# Step 8: Evaluate Model
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Loss:    ", test_loss)
print("Test Accuracy:", test_acc)
```

```python
# Step 9: Make Prediction
predictions = model.predict(x_test)

print("Predicted class:", np.argmax(predictions[0]), '->', class_names[np.argmax(predictions[0])])
print("Actual class:   ", y_test[0], '->', class_names[y_test[0]])
```

```python
# Step 10: Plot Training Loss and Accuracy Graphs
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
